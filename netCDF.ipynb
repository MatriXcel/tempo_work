```python
# Import dependencies. I usually manage Python packages
# using conda.
from netCDF4 import Dataset
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs
import matplotlib.ticker as mticker
import numpy as np
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

# Here I define a generic function to create georefenced maps
def plot_map(ax,x,y,extent,pc,data,vmin=None,vmax=None,title='',cbar_label='',cmap='RdYlBu_r',
             top=False,bottom=False,left=False,right=False,cbar=False,xlocator=None,ylocator=None,alpha=1.0):
    '''
        Use plot_map to create georeferneced plots.
    Input arguments:
        ax (matplotlib axes objec)
        x (float array (1d or 2d)): Center pixel longitudes
        y (float array (1d or 2d)): Center pixel latitudes
        extent (4 float number list setting extent of plot): [min_lon,max_lon,min_lat,max_lat]
        pc (Cartopy projection object)
        data (2d array): Data to be plotted
        vmin (float optional): Minimum colormap value
        vmax (float optional): Maximum colormap value
        title (string optional): Plot title
        cbar_label (string optional): Colorbar label
        top, bottom, left, right (bool optional): Add lat/lon coordinate labels on plot margins logical
        cbar (bool optional): Add colorbar logical
        xlocator (float list optional): Locations for longitude labels
        ylocator (float list optional): Locations for latitude labels
        alpha (float in 0.0-1.0 range optional): map transparency
    Returns:
        None
    '''
    ax.set_title(title)
    ax.set_extent(extent,crs=pc)
    im = ax.imshow(data, origin='upper',extent=(x.min(),x.max(),y.min(),y.max()),transform=pc,
                   interpolation='none',vmin=vmin,vmax=vmax,cmap=cmap,alpha=alpha)
    ax.coastlines(resolution='50m',color='black',linewidth=1)
    ax.add_feature(ccrs.cartopy.feature.STATES)
    gl = ax.gridlines(draw_labels=True,color='black',alpha=0.5)
    gl.xlabels_top=top; gl.xlabels_bottom=bottom; gl.ylabels_right=right; gl.ylabels_left=left
    gl.xformatter  = LONGITUDE_FORMATTER; gl.yformatter = LATITUDE_FORMATTER
    if xlocator:
        gl.xlocator = mticker.FixedLocator(xlocator)
    if ylocator:
        gl.ylocator = mticker.FixedLocator(ylocator)

    if cbar:
        cb = plt.colorbar(im, orientation='horizontal',shrink=0.8, label=cbar_label)

# Define path to netCDF file to open and read
fp = './TEMPO_20130704_17h_LA1_HCHO.nc'

# Open the file. I like to do it using with to guarantee
# that it will be closed after we are done reading
# netCDF files are organized in group (which can have subgroups)
# and data fields. Each group or data field can have attributes.
# The file as a whole can also have attributes that in this case
# are called global attributes. Following this example and reading
# https://unidata.github.io/netcdf4-python/ you should get some
# familiarity about how to work with them
with Dataset(fp,'r') as src:
    # Read geolocation variables

    mirror_step = src['support_data']['amf_molecule_specific'][7]
    x_track = np.arange(61)

    lat = src['geolocation']['latitude'][:]
    lon = src['geolocation']['longitude'][:]
    # Read HCHO AMF
    #col = np.flip(src['support_data']['amf_molecule_specific'][:].T)
    col = np.flip(src['geolocation']['terrain_height'][:].T)

plt.plot(x_track, mirror_step)
# Now plotting. I'm using matplotlib and cartopy to be able
# to produce georefenced plots. I'm defining a function above
# so you can later on adapt and use it in other scripts.
# Set extent
extent=[-120.5,-117.5,33,35] # Something around LA
# Set projection
pc = ccrs.PlateCarree()
# Create image
fig = plt.figure(figsize=(7,7)); fig.suptitle('Test plot')
# Create axe in figure where to plot
ax = fig.add_subplot(1,1,1,projection=pc)
#/1e16
#vmin=2.55,vmax=2.78
#x10$^{16}$ 
plot_map(ax,lon,lat,extent,pc,col,vmin=0,vmax=1.55e3,title='Terrain Height Column',
        top=False,bottom=True,left=True,right=True,cbar=True,cbar_label=' molecules cm$^{-2}$')
        
# Show plot in screen and close it when window is closed.
plt.show(); plt.close()
```
